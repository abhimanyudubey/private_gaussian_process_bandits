```python
import dp_gp as gp
import matplotlib.pyplot as plt

T = 1000
n = 10
d = 5
m = 200
e = gp.Env(n, d)
a = gp.GP_UCB()
b = gp.Random()

reg_a = []
reg_b = []

for t in range(T):
    
    if t % 100 == 0:
        print ('T====', t)
    actions = e.get_action_set()
    
    x_t = a.select_action(actions, beta_mult=0.01)
    y_t, r_t = e.play(x_t)
    a.update(x_t, y_t)
    
    x2_t = b.select_action(actions)
    y2_t, r2_t = e.play(x2_t)
    
    if len(reg_a) > 0:
        reg_a.append(reg_a[-1]+r_t)
        reg_b.append(reg_b[-1]+r2_t)
    else:
        reg_a.append(r_t)
        reg_b.append(r2_t)

plt.plot(reg_a, label='GP-UCB')
plt.plot(reg_b, label='Random')
plt.legend(loc='upper left')
```
